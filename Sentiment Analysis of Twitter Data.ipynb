## Cleaning data

```python
import re
from nltk.tokenize import word_tokenize
from string import punctuation 
from nltk.corpus import stopwords 

import nltk
nltk.download('stopwords')
```

```python
# In this class I will pass the whole dataframe which is not efficient, I have just brought it here for teaching 
class PreProcessTweets:
    def __init__(self):
        self._stopword = set(stopwords.words('english')+list(punctuation)+['AT_USER','URL'])
    def textCleaning(self, row):
        # Note that row is a series, but you can achieve to desired column.
        # Using auxiliary variable is useful.
        a = row['Tweet_text'].lower()
        a= re.sub ( '@apple', 'apple', a)
        a = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', a)
        a = re.sub('@[^\s]+', 'AT_USER', a)
        a=re.sub(r'#([^\s]+)',r'\1', a)
        a = a.split(' ')
        a = [i for i in a if i not in set(stopwords.words('english'))]
        return ' '.join(a)
    def processData(self, dataframe):
        dataframe ['Tweet_text_new'] = dataframe.apply(self.textCleaning , axis=1)
        return dataframe
```

```python
# Here is the efficient solution
class PreProcessTweetsColumn:
    def __init__(self):
        self.stopword = set(stopwords.words('english')+list(punctuation)+['AT_USER','URL'])
    def textCleaning(self, row):
        a = row.lower()
        a= re.sub ( '@apple', 'apple', a)
        a = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', a)
        a = re.sub('@[^\s]+', 'AT_USER', a)
        a=re.sub(r'#([^\s]+)',r'\1', a)
        a = a.split(' ')
        a = [i for i in a if i not in self.stopword]
        return ' '.join(a)
    def processData(self, column_list):
        l=[]
        for i in column_list:
            l.append(self.textCleaning(i))
        return l
```

### Extracting features before splitting

```python
vectorizer=CountVectorizer(min_df=1)
```

```python
X=vectorizer.fit_transform(X_svm).toarray()
```

```python
X
```

```python
vocabulary=vectorizer.get_feature_names()
```

```python
print(len(vocabulary))
vocabulary
```

```python
voc_weights=[]
for word in vocabulary:
    try:
        synset=list(swn.senti_synsets(word))
        meaning =synset[0]
        if meaning.pos_score()>meaning.neg_score():
            weight=meaning.pos_score()
        elif meaning.pos_score()<meaning.neg_score():
            weight=-meaning.neg_score()
        else: 
            weight=0       
        
    except:
        weight=0  

    voc_weights.append(weight)
voc_weights
```

```python
new_X=[]
for i in X:
    new_X.append(np.multiply(i, voc_weights))


    
```
